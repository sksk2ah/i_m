```python
# 모듈 import
import pandas as pd
import yfinance as yf

# ticker를 넣으면 yfinance를 통해 해당하는 티커와 정보를 가져옵니다
def get_ticker_info(ticker):
    ticker_data = yf.Ticker(ticker)
    return ticker_data.info

def main():
    user_input = input("Enter ticker symbol(s) separated by space: ")
    tickers = user_input.split()

    for ticker in tickers:
        ticker_info = get_ticker_info(ticker)
        print(f"Ticker: {ticker}")
        for key, value in ticker_info.items():
            print(f"{key}: {value}")
        print()

if __name__ == "__main__":
    main()
# 주의 - 셀을 실행시키고 티커를 입력하지 않으면 오류가 나거나 셀이 멈추지 않습니다.
```

```python
# 티커에 해당하는 정보를 원하는 것만 볼 수 있게 셀을 따로 뺌
def get_ticker_info(ticker):
    ticker_data = yf.Ticker(ticker)
    return {
        "Ticker": ticker,
        "Recommendation": ticker_data.info.get("recommendationKey"),
        "Open": ticker_data.info.get("open"),
        "Day Low": ticker_data.info.get("dayLow")
    }

def main():
    user_input = input("Enter ticker symbol(s) separated by space: ")
    tickers = user_input.split()

    for ticker in tickers:
        ticker_info = get_ticker_info(ticker)
        print("Ticker:", ticker_info["Ticker"])
        print("Recommendation:", ticker_info["Recommendation"])
        print("Open:", ticker_info["Open"])
        print("Day Low:", ticker_info["Day Low"])
        print()

if __name__ == "__main__":
    main()
```

```python
# S&P 500을 직접적으로 가져올 수 없어 S&P 500 지수를 기반으로 거래되는 ETF가져오기
# SPDR S&P 500 ETF Trust 티커
ticker = "SPY"

# start와 end date 사이의 정보를 가져옵니다.
spy = yf.download(ticker, start="2021-01-01", end="2023-12-31")
spy
```
